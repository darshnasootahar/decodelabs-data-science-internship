```python
# TASK 4: Data Visualization
# DecodeLabs Data Science Internship - Week 1

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# STEP 1: Load Dataset
df = pd.read_csv('/content/Titanic-Dataset.csv')

# Handle missing Age values
df['Age'] = df['Age'].fillna(df['Age'].median())

print("=" * 50)
print("TASK 4: DATA VISUALIZATION")
print("=" * 50)

# Create figure
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

fig.suptitle(
    'Titanic Dataset - Data Visualizations\nDecodeLabs Internship Week 1',
    fontsize=16,
    fontweight='bold'
)

# CHART 1: Survival Count
survival_counts = df['Survived'].value_counts()

axes[0, 0].bar(
    ['Did Not Survive', 'Survived'],
    survival_counts.values,
    color=['red', 'green']
)

axes[0, 0].set_title('1. Survival Count')
axes[0, 0].set_ylabel('Passengers')

# CHART 2: Survival by Gender
gender_survival = df.groupby('Sex')['Survived'].mean() * 100

axes[0, 1].bar(
    gender_survival.index,
    gender_survival.values,
    color=['blue', 'pink']
)

axes[0, 1].set_title('2. Survival Rate by Gender')
axes[0, 1].set_ylabel('Survival Rate (%)')

# CHART 3: Survival by Passenger Class
class_survival = df.groupby('Pclass')['Survived'].mean() * 100

axes[0, 2].bar(
    ['1st Class', '2nd Class', '3rd Class'],
    class_survival.values,
    color=['orange', 'purple', 'cyan']
)

axes[0, 2].set_title('3. Survival Rate by Class')
axes[0, 2].set_ylabel('Survival Rate (%)')

# CHART 4: Age Distribution
axes[1, 0].hist(df['Age'], bins=30, color='skyblue', edgecolor='black')

axes[1, 0].set_title('4. Age Distribution')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Count')

# CHART 5: Fare Distribution
axes[1, 1].hist(df['Fare'], bins=40, color='orange', edgecolor='black')

axes[1, 1].set_title('5. Fare Distribution')
axes[1, 1].set_xlabel('Fare')
axes[1, 1].set_ylabel('Count')

# CHART 6: Embarked Port
embark_counts = df['Embarked'].value_counts()

axes[1, 2].pie(
    embark_counts.values,
    labels=embark_counts.index,
    autopct='%1.1f%%'
)

axes[1, 2].set_title('6. Embarkation Port')

# Adjust layout
plt.tight_layout()

# Save image
plt.savefig('titanic_visualizations.png')

# Show charts
plt.show()

print("✅ All charts created successfully!")
```
