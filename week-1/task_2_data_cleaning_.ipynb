```python
##TASK 2 : Data Cleaning 
# DecodeLabs Data Science Internship - Week 1
import pandas as pd
import seaborn as sns
```

```python
# STEP 1: Load Dataset
df = pd.read_csv('/content/Titanic-Dataset.csv')

print("=" * 50)
print("TASK 2: DATA CLEANING & PREPROCESSING")
print("=" * 50)
```

```python
# STEP 2: Check Missing Values BEFORE
print("\n Missing Values BEFORE Cleaning:")
print(df.isnull().sum())
```

```python
# STEP 3: Handle Missing Values

#Fill 'Age' with median (because age has outliers)
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill 'Embarked' with mode (most common value)
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop 'Cabin' column (too many missing values)
df.drop(columns=['Cabin'], inplace=True)

# STEP 4: Check Missing Values Again
print("\n✅ Missing Values AFTER Cleaning:")
print(df.isnull().sum())

# STEP 5: Display Cleaned Data
print("\nCleaned Dataset:")
print(df.head())
```

```python
# STEP 4: Remove Duplicates
before = df.shape[0]
df.drop_duplicates(inplace=True)
after = df.shape[0]

print(f"\n Duplicates Removed: {before - after}")
print(f"   Rows before: {before}, Rows after: {after}")
```

```python
# STEP 5: First 5 Rows
print("\n  First 5 Rows of Dataset:")
print(df.head())
```

```python
#STEP 5: Format Data Correctly
# ----------------------------

# Convert 'sex' to numeric: male=0, female=1
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Convert 'embarked' to numeric: S=0, C=1, Q=2
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Convert 'survived' and 'pclass' to proper int type
df['Survived'] = df['Survived'].astype(int)
df['Pclass'] = df['Pclass'].astype(int)

print("\n Data Types AFTER Formatting:")
print(df.dtypes)
```

```python
# STEP 6: Final Clean Dataset
print("\n Cleaned Dataset (First 5 Rows):")
print(df.head())

print(f"\n Final Dataset Shape: {df.shape}")

# Save cleaned data
df.to_csv('cleaned_titanic.csv', index=False)
print("\n Cleaned dataset saved as 'cleaned_titanic.csv'")
print(" Task 2 Complete!")
```
