```python
import pandas as pd
import seaborn as sns
```

```python
# STEP 1: Load cleaned Dataset


df = pd.read_csv('/content/cleaned_titanic.csv')

# Fill Age missing values
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked missing values safely
if df['Embarked'].mode().empty:
    print("Embarked column has no mode value")
else:
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop Deck column only if it exists
if 'Deck' in df.columns:
    df.drop(columns=['Deck'], inplace=True)

print(df.head())
print(df.isnull().sum())
print("=" * 50)
print("TASK 3: EXPLORATORY DATA ANALYSIS (EDA)")
print("=" * 50)
```

```python
# STEP 2: Basic Statistics
print("\n  Basic Statistics of Numeric Columns:")
print(df.describe())
```

```python
# STEP 3: Survival Rate
survival_rate = df['Survived'].mean() * 100
print(f"\n Overall Survival Rate: {survival_rate:.2f}%")
print(f"   Survived : {df['Survived'].sum()} Passengers")
print(f"   Died     : {(df['Survived'] == 0).sum()} Passengers")
```

```python
# STEP 4: Survival by Gender
print("\n Survival Rate by Gender:")
gender_survival = df.groupby('Sex')['Survived'].mean() * 100
print(gender_survival.round(2))
```

```python
#STEP 5: Survival by Class
print("\n Survival Rate by Passenger Class:")
class_survival = df.groupby('Pclass')['Survived'].mean() * 100
print(class_survival.round(2))
```

```python
# STEP 6: Age Statistics
print("\n Age Statistics:")
print(f"   Minimum Age : {df['Age'].min()}")
print(f"   Maximum Age : {df['Age'].max()}")
print(f"   Average Age : {df['Age'].mean():.2f}")
print(f"   Median Age  : {df['Age'].median()}")
```

```python
# STEP 7: Fare Statistics
print("\n Fare Statistics:")
print(f"   Minimum Fare : ${df['Fare'].min():.2f}")
print(f"   Maximum Fare : ${df['Fare'].max():.2f}")
print(f"   Average Fare : ${df['Fare'].mean():.2f}")
```

```python
# STEP 8: Outlier Detection
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Fare'] < Q1 - 1.5*IQR) | (df['Fare'] > Q3 + 1.5*IQR)]
print(f"\n  Outliers in 'fare' column: {len(outliers)} Passengers")
```

```python
# STEP 9: Key Findings Summary
print("\n KEY FINDINGS:")
print("""
  1. Overall survival rate was 38.38%
  2. Females had much higher survival rate (~74%) than males (~18%)
  3. First class passengers survived more (~63%) vs Third class (~24%)
  4. Average passenger age was ~29 years
  5. Fare had significant outliers (some paid very high prices)
  6. Most passengers boarded from Southampton port
""")
print("✅ Task 3 Complete!")
```
